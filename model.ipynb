```python
import ast
```

```python
def convert(list):
    res = []
    for i in ast.literal_eval(list):
        res.append(i['name'])
    return res    
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert2(list):
    res = []
    cnt = 0
    
    for i in ast.literal_eval(list):
        if(cnt<3):
            res.append(i['name'])
            cnt += 1
        else:
            break
    return res    
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['crew'][0]
```

```python
def convert3(list):
    res = []
    
    for i in ast.literal_eval(list):
        if i['job'] == 'Director' :
            res.append(i['name'])
            break
    return res       
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
#Overview column contains strings, so we have to convert it also into a list of words
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies_ = movies[['id', 'title', 'tags']]
```

```python
new_movies_
```

```python
new_movies_['tags'] = new_movies_['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies_.head()
```

```python
new_movies_['tags'][0]
```

```python
new_movies_['tags'] = new_movies_['tags'].apply(lambda x:x.lower())
```

```python
new_movies_['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies_['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemm(text):
    res = []
    
    for i in text.split():
        res.append(ps.stem(i))
        
    return " ".join(res)    
```

```python
stemm(new_movies_['tags'][0])
```

```python
new_movies_['tags'] = new_movies_['tags'].apply(stemm)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
similar[0]
```

```python
similar.shape
```

```python
new_movies_[new_movies_['title']=='Avatar']
```

```python
def recommend(movie):
    movie_index = new_movies_[new_movies_['title']==movie].index[0]
    distances = similar[movie_index]
    m_list = sorted(enumerate(distances), reverse=True, key= lambda x:x[1])[1:6]
    
    for i in m_list:
        print(new_movies_.iloc[i[0]].title)
```
